# Part 1 — Data preparation & EDA

Load German electricity demand, aggregate to weekly/daily, plot, decompose, and test for non-stationarity. All heavy lifting lives in `electricity_demand`; this notebook just drives it and shows figures.

```
import warnings; warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from electricity_demand.data import build_processed, load_processed_weekly, load_processed_daily
from electricity_demand import features as feat
from electricity_demand import plotting as plot
```

## Build processed series
Run once (needs `data/raw/time_series_60min_singleindex.csv`). Downloads via `scripts/download_data.py` or drop the file in manually.

```
series = build_processed(save=True)
weekly, daily = series['weekly'], series['daily']
print(weekly.describe())
```

## Weekly and daily plots

```
plot.plot_series(weekly, title='Weekly average German electricity load');
plot.plot_series(daily, title='Daily average German electricity load');
```

## STL decomposition
Look for trend + annual seasonality (period 52 on weekly data).

```
stl = feat.decompose(weekly, period=52)
plot.plot_decomposition(stl);
```

## ACF / PACF
Guides the SARIMA orders in Part 3.

```
plot.plot_acf_pacf(weekly, lags=60);
```

## Stationarity tests (ADF + KPSS)
ADF H0 = unit root (non-stationary); KPSS H0 = stationary. Read them together.

```
print('Level series:');    display(feat.stationarity_report(weekly))
print('First difference:'); display(feat.stationarity_report(weekly.diff().dropna()))
print('Seasonal (52) difference:'); display(feat.stationarity_report(weekly.diff(52).dropna()))
```

**Interpretation.** Comment here on whether the level series is stationary, whether a first and/or seasonal difference is needed, and what that implies for `d`, `D`, and `s=52` in the SARIMA model.
